```python
 # Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

```python
# Count total number of players
uniqueplayercount = len(purchase_data_df["SN"].unique())
total_players_df = pd.DataFrame({"Total Players" : [uniqueplayercount]})
total_players_df
```

```python
# Count total number of unique items
uniqueitemcount = len(purchase_data_df["Item ID"].unique())

#Calculate average price
averageprice = '${:,.2f}'.format(purchase_data_df["Price"].mean())

# Count total number of purchases
totalpurchases = purchase_data_df["Item ID"].count()

#Calculate total revenue
totalrev = '${:,.2f}'.format(purchase_data_df["Price"].sum())

#Create summary dataframe
purchases_summary_df = pd.DataFrame({"Number of Unique Items" : uniqueitemcount, "Average Price" : [averageprice], "Number of Purchases" : [totalpurchases], "Total Revenue" : [totalrev]})
purchases_summary_df
```

```python
#Data Frame grouped by gender
gender_purchases_df = purchase_data_df.groupby(['Gender'])

#Count the number of unique players 
gendercount = gender_purchases_df["SN"].nunique()

#Calculate Percentage of players from each gender
genderpercent = (gendercount/uniqueplayercount)*100

#Create summary dataframe
gender_demographics_df = pd.DataFrame({"Total Count": gendercount,
                                    "Percentage of Players": genderpercent})

#Format perecent of players column
gender_demographics_df["Percentage of Players"] = gender_demographics_df["Percentage of Players"].astype(float).map("{:,.2f}%".format)

gender_demographics_df
```

```python
#Calculate purchase count
purchasecount = gender_purchases_df["Purchase ID"].count()

#Calculate average purchase price
avgpurchaseprice = gender_purchases_df["Price"].mean().astype(float).map("${:,.2f}".format)

#Calculate Total Purchase Value
totalpurchasevalue = gender_purchases_df["Price"].sum().astype(float).map("${:,.2f}".format)

#Calculate Total Purchase per Person
avgpurchasepp = (gender_purchases_df["Price"].sum()) / (gender_purchases_df["SN"].nunique())

#Create summary dataframe
gender_purchasing_df = pd.DataFrame({"Purchase Count": purchasecount,
                                    "Average Purchase Price": avgpurchaseprice,
                                    "Total Purchase Value": totalpurchasevalue,
                                    "Avg Total Purchase per Person": avgpurchasepp})

#Format Avg Total Purchase per Person column
gender_purchasing_df["Avg Total Purchase per Person"] = gender_purchasing_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

gender_purchasing_df
```

```python
#Find max age
purchase_data_df["Age"].max()

# Create the bins
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]

# Create labels for these bins
age_groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Slice the data and place it into bins
purchase_data_df["Age Groups"] = pd.cut(purchase_data_df["Age"], bins, labels=age_groups, include_lowest=True)

#Group data by age groups
age_groups_df = purchase_data_df.groupby(['Age Groups'])

#Count of age groups
agegroupcount = age_groups_df["SN"].nunique()

#Percentage of Players 
agegrouppercent = (agegroupcount/uniqueplayercount) * 100

#Create summary dataframe
age_demographics_df = pd.DataFrame({"Total Count": agegroupcount,
                                    "Percentage of Players": agegrouppercent})

#Format Avg Total Purchase per Person column
age_demographics_df["Percentage of Players"] = age_demographics_df["Percentage of Players"].astype(float).map("{:,.2f}%".format)

age_demographics_df
```

```python
#Calculate purchase count
agepurchasecount = age_groups_df["Purchase ID"].count()

#Calculate average purchase price
ageavgpurchaseprice = age_groups_df["Price"].mean().astype(float).map("${:,.2f}".format)

#Calculate Total Purchase Value
agetotalpurchasevalue = age_groups_df["Price"].sum().astype(float).map("${:,.2f}".format)

#Calculate Total Purchase per Person
ageavgpurchasepp = (age_groups_df["Price"].sum()) / (age_groups_df["SN"].nunique())

#Create summary dataframe
age_purchasing_df = pd.DataFrame({"Purchase Count": agepurchasecount,
                                    "Average Purchase Price": ageavgpurchaseprice,
                                    "Total Purchase Value": agetotalpurchasevalue,
                                    "Avg Total Purchase per Person": ageavgpurchasepp})

#Format Avg Total Purchase per Person column
age_purchasing_df["Avg Total Purchase per Person"] = age_purchasing_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

age_purchasing_df
```

```python
#Group data by SN
SN_df = purchase_data_df.groupby(['SN'])

#Purchase count
SNpurchasecount = SN_df["Purchase ID"].count()

#Average Purchase Price
SNpurchaseavg = SN_df["Price"].mean().astype(float).map("${:,.2f}".format)

#Total Purchase Value
SNpurchasesum = SN_df["Price"].sum().astype(float).map("${:,.2f}".format)

#Create summart data frame
top_spenders_df = pd.DataFrame({"Purchase Count": SNpurchasecount,
                                    "Average Purchase Price": SNpurchaseavg,
                                    "Total Purchase Value": SNpurchasesum})

#Sort by purachse count decending
sortedtop_spenders_df = top_spenders_df.sort_values("Purchase Count", ascending=False)
sortedtop_spenders_df
```

```python
#Group data by Item ID and Name
item_df = purchase_data_df.groupby(['Item ID', 'Item Name'])

#Purchase count
itempurchasecount = item_df["Purchase ID"].count()

#Item Price
itempurchaseprice = item_df["Price"].mean().astype('int64').map("${:,.2f}".format)

#Total Purchase Value
itempurchasesum = item_df["Price"].sum().astype('int64').map("${:,.2f}".format)

#Create summart data frame
popular_items_df = pd.DataFrame({"Purchase Count": itempurchasecount,
                                    "Item Purchase Price": itempurchaseprice,
                                    "Total Purchase Value": itempurchasesum})

#Sort by purachse count decending
sortedpopular_items_df = popular_items_df.sort_values("Purchase Count", ascending=False)
sortedpopular_items_df
```
